# ED: Project 2 - Model Training

After the data were all porcessed and normalize so that the models used don't have problems when trying to find the necessary correlation, if present, the following step is to chose the models themshelves to use and to train them.
We decided to chose only supervised machine learning algorithm, settling on the following one:
* Support Vector Machine
* Mulilayer Perceptron
* Decision Tree
* Random Forest
* K Nearest Neighbors

The first step is obviously to load the dataset to memory. 

```python
import pandas as pd
myData = pd.read_csv("ourDataset.csv")
myData
```

## Test and Train Division

The second step is to define which portion of the dataset that we want to be used as training data and the portion that we want to be used as testing data. In general, this division is made according to the dataset size, since we want to give as much data as possible to the model during the training phase. In our case we opted to can a 80/20 division, assigning 80% of the data to the training phase and the remaining for the testing phase.

```python
#Dataset Division

from sklearn.model_selection import train_test_split

training_set, test_set = train_test_split(myData, test_size = 0.2, random_state = 1)

X_train = training_set.iloc[:,0:-2].values
Y_train = training_set.iloc[:,-1].values
X_test = test_set.iloc[:,0:-2].values
Y_test = test_set.iloc[:,-1].values
```

## Support Vector Machine

This section is the implementation of the SVM model, the calculation of its prediction as well as the chosen performance metrics(Accuracy and F-Score).

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

from sklearn.svm import SVC

clrSVM = SVC(kernel='linear', random_state = 1, gamma="auto")
clrSVM.fit(X_train,Y_train)

predSVM = clrSVM.predict(X_test)

#cmSVM = confusion_matrix(Y_test,predictions)
print("Accuracy for Support Vector Machine Classifier: " + str(accuracy_score(Y_test, predSVM)*100)+"%")
print("F1-Score for Support Vector Machine Classifier: " + str(f1_score(Y_test, predSVM)*100)+"%")
```

## MultiLayer Perceptron

This section is the implementation of the MLP model, the calculation of its prediction as well as the chosen performance metrics(Accuracy and F-Score).

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier

scaler = StandardScaler()
scaler.fit(X_train)

X_trainMLP = scaler.transform(X_train)
X_testMLP = scaler.transform(X_test)

clrMLP = MLPClassifier(hidden_layer_sizes=(10,5),max_iter=900,activation="identity",solver="lbfgs")
clrMLP.fit(X_trainMLP,Y_train)

predNN = clrMLP.predict(X_testMLP)

#cmNN = confusion_matrix(Y_test,predictions)
print("Accuracy for Neural Network Classifier: " + str(accuracy_score(Y_test, predNN)*100)+"%")
print("F1-Score for Neural Network Classifier: " + str(f1_score(Y_test, predNN)*100)+"%")
```

## Decision Tree

This section is the implementation of the DP model, the calculation of its prediction as well as the chosen performance metrics(Accuracy and F-Score).

```python
from sklearn.tree import DecisionTreeClassifier

clrDT = DecisionTreeClassifier(criterion="gini")
clrDT = clrDT.fit(X_train, Y_train)

predDT = clrDT.predict(X_test)

#cmDT = confusion_matrix(Y_test,predictions)
print("Accuracy for Decision Tree Classifier: " + str(accuracy_score(Y_test, predDT)*100)+"%")
print("F1-Score for Decision Tree Classifier: " + str(f1_score(Y_test, predDT)*100)+"%")
```

## Random Forest

This section is the implementation of the RF model, the calculation of its prediction as well as the chosen performance metrics(Accuracy and F-Score).

```python
from sklearn.ensemble import RandomForestClassifier

clrRF = RandomForestClassifier(n_estimators=15)
clrRF = clrRF.fit(X_train, Y_train)

predRF = clrRF.predict(X_test)

#cmRF = confusion_matrix(Y_test,predictions)
print("Accuracy for Random Forrest Classifier: " + str(accuracy_score(Y_test, predRF)*100)+"%")
print("F1-Score for Random FOrrest Classifier: " + str(f1_score(Y_test, predRF)*100)+"%")
```

## K Nearest Neighbors

This section is the implementation of the KNN model, the calculation of its prediction as well as the chosen performance metrics(Accuracy and F-Score).

```python
from sklearn.neighbors import KNeighborsClassifier

clrKN = KNeighborsClassifier(n_neighbors=3)
clrKN = clrKN.fit(X_train, Y_train)

predKN = clrKN.predict(X_test)

#cmKN = confusion_matrix(Y_test,predictions)
print("Accuracy for KNeighbors Classifier: " + str(accuracy_score(Y_test, predKN)*100)+"%") 
print("F1-Score for KNeighbors Classifier: " + str(f1_score(Y_test, predKN)*100)+"%")
```
